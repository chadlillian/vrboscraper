```python
import pandas as pd
import matplotlib.pyplot as plt
import folium
from pymongo import MongoClient
```

```python
DATABASE = 'VRBO_scenic_hwy_98'
DATABASE = '53odessa'
client = MongoClient()
db = client[DATABASE]
print db.collection_names()
stats = db['Stats']
calendars = db['Calendar']
```

```python
a = stats.find()
coords = []
for aa in a:
    if 'lng' in aa.keys():
        coords.append([aa['lat'], aa['lng']])
print len(coords)
```

```python
print coords[0]
```

```python
radars = folium.Map(location=coords[0], zoom_start=17,tiles='OpenStreetMap', width=800, height=600)
for c in coords:
    #radars.simple_marker(location=c)
    radars.circle_marker(location=c, radius=5,
                    popup='comp', line_color='#3186cc',
                    fill_color='#3186cc')
```
